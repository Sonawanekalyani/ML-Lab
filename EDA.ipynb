```python

# Function for log transformation of the column
def log_transform(data, col):
    for colname in col:
        # Check if column contains only 1.0 values
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname] + 1)
        else:
            # Avoid log(0) or log of negative numbers by checking the minimum value
            if (data[colname] <= 0).any():
                print(f"Warning: Column '{colname}' contains non-positive values. Log transformation skipped for this column.")
                data[colname + '_log'] = np.nan  # You can choose how to handle this case
            else:
                data[colname + '_log'] = np.log(data[colname])
    
    data.info()
```
